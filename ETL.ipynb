### Orders Table

```python
df_return = pd.read_csv('return.csv')
```

```python
df_return.head()
```

```python
order_table = df[['Order ID', 'Customer ID','Postal Code','Ship Mode','OrderDate','ShipDate']]
```

```python
order_table
```

```python
joindata = order_table.merge(df_return , on = ['Order ID'])
```

```python
joindata
```

```python
list_order_return = joindata['Order ID'].unique()
```

```python
list_order_return
```

```python
order_table[['Return']] = False
```

```python
for i in order_table.index:
    if order_table['Order ID'][i] in list_order_return:
        order_table['Return'][i] = True
    else:
        order_table['Return'][i] = False
```

```python
order_table
```

```python
order_table = order_table.replace({'Ship Mode': {'Second Class': 1, 'Standard Class': 2,'First Class': 3,'Same Day': 4}})
```

### Metatable

```python
import datetime
```

```python
now = datetime.date.today()
```

```python
df_return = pd.read_csv('return.csv')
```

```python
return_list = ['bad condition', 'wrong item', 'wanted to returned']
prod = df_return['Order ID'].unique()
prod = pd.DataFrame(prod, columns=['Order ID'])
prod['Reason'] = np.nan
prod['Date'] = now
prod['Conditions'] = 'Good'
for i in prod.index:
    returned = random.choice(return_list)
    prod['Reason'][i] = returned
```

```python
metatable = df[['Customer ID','Order ID','Region']]
```

```python
metatable
```

```python
metatable = metatable.merge(prod,on='Order ID')
```

```python
metatable = metatable.replace({'Region': {'West':1 , 'East':2,'Central':3,'South':4}})
```

```python
metatable = metatable.drop_duplicates()
```

```python
metatable.to_csv('metadata.csv')
```
